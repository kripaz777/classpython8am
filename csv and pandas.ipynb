```python
import csv
csv_columns = ['name', 'age', 'add']
data = [
    {'name': 'Aayush', 'age': '18', 'add': 'Kathmandu'},
    {'name': 'Shyam', 'age': '23', 'add': 'Bhaktapur'},
    {'name': 'Hari', 'age': '45', 'add': 'Lalitpur'}
]

csvfile = open("data.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
writer.writeheader()
writer.writerows(data)
csvfile.close()
```
